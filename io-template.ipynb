## Reading and writing files

```python
# put the filename – and path if necessary – in a variable
filename = "../data/covid_genome/covid.fasta"

# open the file
genome = open(filename, "r")

# we created a variable ('genome') when we opened the file: what is it?
print(type(genome))
print(genome)
```

### Reading a file with `read()`

```python
# we interact with the file through the file handle, called 'genome'

# read the file
# read() reads in the entire file in one variable, in this case 'dna_sequence'
dna_sequence = genome.read()
print(dna_sequence)
```

```python
# close the file
genome.close()
```

```python
# let's read the file again
dna = genome.read()
```

### Opening a file with `with()`

```python
# another way to read a file
with open("../data/fruit.txt") as fruit:
    fruit_data = fruit.read()
```

```python
print('----begin----')
print(fruit_data)
print('----end----')
```

### looping over the lines in a file

```python
# use a for() loop to read the file line by line
input_file = "../data/cowboy_carter.txt"
album = open(input_file, "r")


# make a counter for the track number 
track_num = 1

for track in album:
    # one way to tget rid of the extra line break is to strip it off
    track = track.rstrip()
    print(f"track number {track_num} is {track}")

    # one way to get rid of the extra line break is end you rpint with end =''
    # print(f"track number {track_num} is {track}", end='')
    track_num += 1
```

```python
# close the output file
album.close()
```
