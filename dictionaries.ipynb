```python
from collections import defaultdict
import csv

# declare some dictionaries
presidents = defaultdict(str) # key = president number, value = president name
party_count = defaultdict(int) # key = party, value = count of presidents in that party
# without defaultdict, we'd declare a dictionary like this: parties = {}
# party_count = {}

# open the presidents.csv data file 
input_file = "../data/presidents.csv"

# open the file
with open(input_file, 'r') as president_file:
    # create a csv reader object
    reader = csv.reader(president_file, delimiter=',')

    # skip the header line
    header = next(reader)

    # read the file line by line
    for line in reader:
        # print just to make sure we are reading the file right
        # print(line)

        # strip the line breaks <-- depracated
        # line = line.strip()

        # split each line into a list <-- depracated and replacing with csv reader
        # columns = line.split(',')

        # did we split it?
        # print(line[0])

        # load our presidents{} dictionary
        num = line[0].strip()
        name = line[1].strip()
        presidents[num] = name
        # ------- OR ------- 
        # presidents[columns[0]] = columns[1]

        # load the party_count dictionary
        party = line[5].strip()
        party_count[party] += 1

```

```python
# get stuff from our dictionary 
print(presidents.get('26'))
print(presidents['26'])
print(presidents.get('48'), "Not elected yet")
```

```python
# loop over the dictionary
for num in presidents.keys():
    print(f'president number {num} was {presidents.get(num)}')
```

```python
# instead of printing, let's write our output to a file
outfile = open("pres.out", "w")

# another way to loop over a dictionary
for num, name in presidents.items():
    outfile.write(f'president number {num} was {name}\n')
```

```python
# data doesn't get written until you close it
outfile.close()
```
